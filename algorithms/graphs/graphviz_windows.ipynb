Basically, the Python libraries used to do this kind of operation are actually wrappers/interfaces around a popular Open Source tool called [Graphviz](https://www.graphviz.org/), so merely `pip install`ing your library of choice only gets at the wrapper, not the underlying codebase that'll do the heavy lifting.

Because we want to install new Python utilities that rely on non-python-file/module backends, it's most appropriate to leverage Anaconda. The following command will get you need:

```
conda install graphviz pydot
```
